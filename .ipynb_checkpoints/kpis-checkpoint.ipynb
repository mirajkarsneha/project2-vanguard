# KPI 1: Completion Rate: The proportion of users who reach the final ‘confirm’ step

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
import scipy.stats as st
from scipy.stats import ttest_ind, norm

# Load your data
df_final_experiment_clients = pd.read_csv('DataFiles/final_data.csv')

df_final_experiment_clients['Variation'] = df_final_experiment_clients['Variation'].str.strip().str.lower()
clients_started = df_final_experiment_clients['client_id'].nunique()

clients_reached_confirm = df_final_experiment_clients[df_final_experiment_clients['process_step'] == 'confirm']['client_id'].nunique()

#overall_completion_rate = (clients_reached_confirm / clients_started) * 100
#print(f"Overall Completion Rate: {overall_completion_rate:.2f}%")

test_group_clients = df_final_experiment_clients[df_final_experiment_clients['Variation'] == 'test']
test_group_clients.to_csv('DataFiles/test_group.csv', index=False)

control_group_clients = df_final_experiment_clients[df_final_experiment_clients['Variation'] == 'control']
control_group_clients.to_csv('DataFiles/control_group.csv', index=False)

test_group_completion = test_group_clients[test_group_clients['process_step'] == 'confirm']['client_id'].nunique()
control_group_completion = control_group_clients[control_group_clients['process_step'] == 'confirm']['client_id'].nunique()

test_completion_rate = (test_group_completion / test_group_clients['client_id'].nunique()) * 100
control_completion_rate = (control_group_completion / control_group_clients['client_id'].nunique()) * 100

difference_completion_rate = test_completion_rate - control_completion_rate

print(f"Test Group Completion Rate: {test_completion_rate:.2f}%")
print(f"Control Group Completion Rate: {control_completion_rate:.2f}%")
print(f"Difference Completion Rate: {difference_completion_rate:.2f}%")

# Create a DataFrame for plotting
data = pd.DataFrame({
    'Group': ['Test Group', 'Control Group'],
    'Completion Rate (%)': [round(test_completion_rate, 2), round(control_completion_rate, 2)]
})

df_summary = pd.DataFrame(data)
df_summary.to_csv('DataFiles/kpi1_completion_rates.csv', index=False)

plt.figure(figsize=(8, 6))
sns.barplot(x='Group', y='Completion Rate (%)', data=data, palette='Blues')
for index, row in data.iterrows():
    plt.text(index, row['Completion Rate (%)'] + 1, f'{row["Completion Rate (%)"]:.2f}%', ha='center')
plt.show()
```

# KPI 2 : Time Spent on Each Step

```python
##################### KPI 2 - test Group- Time Spent on Each Step: The average duration users spend on each step ##########################
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import ttest_ind
import scipy.stats as st
from scipy.stats import ttest_ind, norm

test_group_clients = pd.read_csv('DataFiles/test_group.csv')
control_group_clients = pd.read_csv('DataFiles/control_group.csv')

test_group_clients['date_time'] = pd.to_datetime(test_group_clients['date_time'])
test_group_clients = test_group_clients.sort_values(by=['client_id', 'date_time'])

test_group_clients['date_time'] = test_group_clients.groupby('client_id')['date_time'].diff()
test_group_average_time_per_step = test_group_clients.groupby('process_step')['date_time'].mean()
 
print("Test Group-Average Time Spent on Each Step:")
print(test_group_average_time_per_step)

##################### KPI 2 - Control Group- Time Spent on Each Step: The average duration users spend on each step ##########################

control_group_clients['date_time'] = pd.to_datetime(control_group_clients['date_time'])
control_group_clients = control_group_clients.sort_values(by=['client_id', 'date_time'])

control_group_clients['date_time'] = control_group_clients.groupby('client_id')['date_time'].diff()
control_group_average_time_per_step = control_group_clients.groupby('process_step')['date_time'].mean()

# Print the results
print("Control Group-Average Time Spent on Each Step:")
print(control_group_average_time_per_step)

# Assuming test_group_average_time_per_step and control_group_average_time_per_step are already defined
new_order = ['start', 'step_1', 'step_2', 'step_3', 'confirm']

test_group_average_time_per_step = test_group_average_time_per_step.reindex(new_order).fillna(0)
control_group_average_time_per_step = control_group_average_time_per_step.reindex(new_order).fillna(0)

data = pd.DataFrame({
    'Step': new_order,
    'Test Group (Avg Time)': test_group_average_time_per_step.values,
    'Control Group (Avg Time)': control_group_average_time_per_step.values
})

df_summary1 = pd.DataFrame(data)
df_summary1.to_csv('DataFiles/kpi2_time_spent.csv', index=False)
print(df_summary1)

bar_width = 0.35
plt.figure(figsize=(12, 6))
x = np.arange(len(new_order))

plt.bar(x - bar_width/2, test_group_average_time_per_step, width=bar_width, label='Test Group', alpha=0.7)
plt.bar(x + bar_width/2, control_group_average_time_per_step, width=bar_width, label='Control Group', alpha=0.7)

plt.xlabel('Process Step')
plt.ylabel('Average Time (Seconds)')
plt.title('Average Time Spent on Each Step')
plt.xticks(x, new_order)  # Set x-ticks to the process step names
plt.legend()
plt.show()
```

# KPI 3 : Error Rates

```python
# If there’s a step where users go back to a previous step, it may indicate confusion or an error. 
# You should consider moving from a later step to an earlier one as an error
df_final_experiment_clients['Variation'] = df_final_experiment_clients['Variation'].str.strip().str.lower()

step_order = ['start', 'step_1', 'step_2', 'step_3', 'confirm']

step_order_dict = pd.Series(range(len(step_order)), index=step_order).to_dict()

def count_backward_movements(steps):
    errors = 0
    previous_index = -1 

    for step in steps:
        current_index = step_order_dict[step] 
        if current_index < previous_index:
            errors += 1
        previous_index = current_index

    return errors
    
# Calculate total users in each group
test_group_total_users = test_group_clients['client_id'].nunique()
control_group_total_users = control_group_clients['client_id'].nunique()    
    
# Calculate test group errors
test_group_error = test_group_clients.groupby(['client_id', 'visit_id'])['process_step'].apply(list)
test_group_error = test_group_error.apply(count_backward_movements)
test_group_total_errors = test_group_error.sum()
print(f"Test group errors: {test_group_total_errors}")

# Calculate control group errors
control_group_error = control_group_clients.groupby(['client_id', 'visit_id'])['process_step'].apply(list)
control_group_error = control_group_error.apply(count_backward_movements)
control_group_total_errors = control_group_error.sum()  
print(f"Control group errors: {control_group_total_errors}")

# Calculate total errors
df_final_experiment_errors = df_final_experiment_clients.groupby(['client_id', 'visit_id'])['process_step'].apply(list).apply(count_backward_movements)
total_errors = df_final_experiment_errors.sum()
print(f"Total backward movements/errors for all clients: {total_errors}")

# Optionally, show errors for each client
print("Errors per client in overall experiment:")
print(df_final_experiment_errors[df_final_experiment_errors > 0])

def calculate_error_rate(error_counts, total_users):
    return (error_counts / total_users) * 100 if total_users > 0 else 0

# Calculate error rates
test_group_error_rate = calculate_error_rate(test_group_total_errors, test_group_total_users)
control_group_error_rate = calculate_error_rate(control_group_total_errors, control_group_total_users)

# Print error rates
print(f"Test Group Error Rate: {test_group_error_rate:.2f}%")
print(f"Control Group Error Rate: {control_group_error_rate:.2f}%")

# Create a DataFrame for plotting
data = pd.DataFrame({
    'Group': ['Test Group', 'Control Group'],
    'Errors Rate': [f"{round(test_group_error_rate, 2)}%",f"{round(control_group_error_rate, 2)}%" ]

})

df_summary2 = pd.DataFrame(data)
df_summary2.to_csv('DataFiles/kpi3_error_rates.csv', index=False)
print(df_summary2)

# Bar plot 
labels = ['Test Group', 'Control Group']
error_counts = [test_group_error_rate, control_group_error_rate]

plt.bar(labels, error_counts, color=['blue', 'orange'])
for index, count in enumerate(error_counts):
    plt.text(index, count + 1, f'{count}', ha='center', va='bottom')
plt.show()
```

# Hypothesis 1 : Completion Rate

#### H0: There is no significant difference in the completion rates between the Test and Control groups.
#### H1: The completion rate for the Test group is significantly higher than that of the Control group.

```python
# Prepare the data for T-test
test_completions = np.array(test_group_clients['process_step'] == 'confirm', dtype=int)
control_completions = np.array(control_group_clients['process_step'] == 'confirm', dtype=int)*1.05

# Perform T-test
t_stat, p_value = ttest_ind(test_completions, control_completions, equal_var = False, alternative = "greater")

print(f"T-Statistic: {t_stat:.4f}, P-Value: {p_value:.4e}")

# Hypothesis Testing
alpha = 0.05

if p_value < alpha:
    print("Reject the null hypothesis. The completion rate for the Test group is significantly different from the Control group.")
else:
    print("Fail to reject the null hypothesis. There is no significant difference in completion rates between the Test and Control groups.")
```

# Hypothesis 2 : Completion Rate with a Cost-Effectiveness Threshold

#### H0: The new UI design meets the cost-effectiveness threshold.
#### H1: The new UI design does NOT meet the cost-effectiveness threshold.

```python
# Calculate the difference in completion rates
rate_difference = test_completion_rate - control_completion_rate

# Check if the difference meets the 5% threshold
threshold = 5

# Check if the increase meets the threshold
if rate_difference >= threshold:
    result = "The new UI design meets the cost-effectiveness threshold."
else:
    result = "The new UI design does NOT meet the cost-effectiveness threshold."

print(f"Test Group Completion Rate: {test_completion_rate:.2f}%")
print(f"Control Group Completion Rate: {control_completion_rate:.2f}%")
print(f"Increase in Completion Rate: {rate_difference:.2f}%")
print(result)
```

# Hypothesis 3: Average age of clients

#### H0 : The average ages are significantly different
#### H1 : The average ages are not significantly different

```python
# You might want to test whether the average age of clients engaging with the new process is the same as those engaging with the old process
import numpy as np
import scipy.stats as stats
import pandas as pd

# Load your data
df_final_experiment_clients = pd.read_csv('DataFiles/final_data.csv')

# Extract ages for test and control groups
test_group_ages = df_final_experiment_clients[df_final_experiment_clients['Variation'] == 'Test']['clnt_age']
control_group_ages = df_final_experiment_clients[df_final_experiment_clients['Variation'] == 'Control']['clnt_age']

# Calculate means
mean_test = np.mean(test_group_ages)
mean_control = np.mean(control_group_ages)

# Perform t-test
t_statistic, p_value = stats.ttest_ind(test_group_ages, control_group_ages, equal_var=False)

# Significance level
alpha = 0.05

# Print results
print(f"Mean Age (Test Group): {mean_test:.2f}")
print(f"Mean Age (Control Group): {mean_control:.2f}")
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4e}")

# Decision based on p-value
if p_value < alpha:
    result = "Reject the null hypothesis: The average ages are not significantly different."
else:
    result = "Fail to reject the null hypothesis: The average ages are significantly different."

print(result)
```

# Hypothesis 4: The average client tenure

#### H0: The average client tenure of those engaging with the new process is equal to the average tenure of those engaging with the old process.
#### H1: The average client tenure of those engaging with the new process is not equal to the average tenure of those engaging with the old process.

```python
import numpy as np
import scipy.stats as stats
import pandas as pd

# Load your data
df_final_experiment_clients = pd.read_csv('DataFiles/final_data.csv')

test_group_clients = df_final_experiment_clients['clnt_tenure_yr']
control_group_clients = df_final_experiment_clients['clnt_tenure_yr']

# Perform T-test
t_stat, p_value = ttest_ind(test_group_clients, control_group_clients, alternative = "two-sided", equal_var=False)
mean_test = np.mean(test_group_clients)
mean_control = np.mean(control_group_clients)

alpha = 0.05

print(f"Mean Age (Test Group): {mean_test:.2f}")
print(f"Mean Age (Control Group): {mean_control:.2f}")
print(f"T-statistic: {t_statistic:.4f}")
print(f"P-value: {p_value:.4e}")

if p_value < alpha:
    result = "Reject the null hypothesis: The average tenures are significantly different."
else:
    result = "Fail to reject the null hypothesis: The average tenures are not significantly different."

print(result)
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the datasets
test_group_clients = pd.read_csv('DataFiles/test_group.csv')
control_group_clients = pd.read_csv('DataFiles/control_group.csv')

# Reset index to avoid duplicate index labels
test_group_clients = test_group_clients.reset_index(drop=True)
control_group_clients = control_group_clients.reset_index(drop=True)

# 1. Age Group Distribution
age_bins = [18, 25, 35, 45, 55, 65, 75, 85, 95]
age_labels = ['18-24', '25-34', '35-44', '45-54', '55-64', '65-74', '75-84', '85+']
test_group_clients['age_group'] = pd.cut(test_group_clients['clnt_age'], bins=age_bins, labels=age_labels)
control_group_clients['age_group'] = pd.cut(control_group_clients['clnt_age'], bins=age_bins, labels=age_labels)

# Combine the datasets and add a 'Group' column
test_group_clients['Group'] = 'Test Group'
control_group_clients['Group'] = 'Control Group'
combined_data = pd.concat([test_group_clients, control_group_clients], ignore_index=True)

# Plot Age Group Distribution
plt.figure(figsize=(12, 6))
sns.countplot(data=combined_data, x='age_group', hue='Group', order=age_labels)
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.title('Age Group Distribution: Test vs Control Group')
plt.legend(title='Group')
plt.show()

# 2. Client Tenure Analysis
tenure_bins = [0, 1, 5, 10, 20, 50]
tenure_labels = ['Less than 1 year', '1-5 years', '5-10 years', '10-20 years', '20+ years']
combined_data['tenure_group'] = pd.cut(combined_data['clnt_tenure_yr'], bins=tenure_bins, labels=tenure_labels)

# Plot Tenure Group Distribution
plt.figure(figsize=(12, 6))
sns.countplot(data=combined_data, x='tenure_group', hue='Group', order=tenure_labels)
plt.xlabel('Tenure Group')
plt.ylabel('Count')
plt.title('Tenure Group Distribution: Test vs Control Group')
plt.legend(title='Group')
plt.show()

# 3. Gender Distribution
plt.figure(figsize=(8, 6))
sns.countplot(data=combined_data, x='gender', hue='Group')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Gender Distribution: Test vs Control Group')
plt.legend(title='Group')
plt.show()
```
